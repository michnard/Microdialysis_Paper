```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import RTs.csv as a dataframe
data = pd.read_csv('../../data/raw/lcms/RTs.csv', header=None)
# # turn nans into zeros
data = data.fillna(0)
# # merge first and second column
data.loc[:,0] = data.loc[:,0] + data.loc[:,1]
data = data.drop(columns=[1])
```

```python
# plot three histograms of the RTs in the dataframe
plt.figure(figsize=(3, 6))
plt.subplot(311)
plt.hist(np.log(data.loc[:,0]), bins=np.arange(-4, 6, 0.5), alpha=0.5, color='grey',label = 'Mouse 1')
plt.xlabel('log RTs error')
plt.ylabel('Count')
plt.legend()
plt.subplot(312)
plt.hist(np.log(data.loc[:,2]), bins=np.arange(-4, 6, 0.5), alpha=0.5, color='grey',label = 'Mouse 2')
plt.xlabel('log RTs error')
plt.ylabel('Count')
plt.legend()
plt.subplot(313)
plt.hist(np.log(data.loc[:,3]), bins=np.arange(-4, 6, 0.5), alpha=0.5, color='grey',label = 'Mouse 3')
plt.xlabel('log RTs error')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.savefig('../../figures/supp/RTs_histogram.pdf', dpi=300)
```
