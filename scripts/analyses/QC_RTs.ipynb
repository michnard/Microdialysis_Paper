```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import RTs.csv as a dataframe

data = pd.read_csv('RTs.csv', header=None)
data
```

```python
data.loc[:,0]
```

```python
# turn nans into zeros
data = data.fillna(0)
# merge first and second column
data.loc[:,0] = data.loc[:,0] + data.loc[:,1]
```

```python
data
```

```python
# plot three histograms of the RTs in the dataframe
plt.figure(figsize=(3, 6))
plt.subplot(311)
plt.hist(np.log(data.loc[:,0]), bins=np.arange(-4, 6, 0.5), alpha=0.5, color='b',label = 'Mouse 1')
plt.xlabel('log RTs error')
plt.ylabel('Count')
plt.legend()
plt.subplot(312)
plt.hist(np.log(data.loc[:,2]), bins=np.arange(-4, 6, 0.5), alpha=0.5, color='g',label = 'Mouse 2')
plt.xlabel('log RTs error')
plt.ylabel('Count')
plt.legend()
plt.subplot(313)
plt.hist(np.log(data.loc[:,3]), bins=np.arange(-4, 6, 0.5), alpha=0.5, color='r',label = 'Mouse 3')
plt.xlabel('log RTs error')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.savefig('RTs_histogram.pdf', dpi=300)
```
